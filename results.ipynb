```python
import matplotlib.pyplot as plt
import numpy as np

# Generate example data
models = ['BERT', 'ENSEMBLE', 'ENTROPY']
metrics = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc', 'log_loss', 'mcc']
scores = {
    'BERT': [0.96, 0.99, 0.95, 0.96, 0.98, 0.21, 0.90],
    'ENSEMBLE': [0.94, 0.99, 0.92, 0.95, 0.98, 0.23, 0.89],
    'ENTROPY': [0.61, 0.61, 1, 0.76, 0.8, 0.65, 0.30]
}
# Define colors and styles
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
hatch_patterns = ['/', 'o', 'x']

# Create subplots
fig, axs = plt.subplots(len(metrics), 1, figsize=(10, len(metrics) * 5))

# Plot each metric for each model
for i, metric in enumerate(metrics):
    for j, model in enumerate(models):
        axs[i].bar(model, scores[model][i], color=colors[j], hatch=hatch_patterns[j], label=model)

    axs[i].set_ylabel(metric.capitalize())
    axs[i].set_ylim(0, 1)  # Assuming scores are between 0 and 1
    axs[i].set_title(f'{metric.capitalize()} Comparison')
    axs[i].legend()

plt.tight_layout()
plt.show()
```
